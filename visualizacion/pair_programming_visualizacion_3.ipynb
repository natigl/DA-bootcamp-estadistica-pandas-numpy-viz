**EJERCICIO 4**

Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el n칰mero de ataques por pa칤s y a침o desde el a침o 1900.

```python
# Filtramos por los paises, expecificando en el m칠todo .isin() cuales de ellos.

df_paises =df[df['country'].isin(['USA','AUSTRALIA', 'SOUTH AFRICA'])]
df_paises.head(2)
```

```python
# Volvemos a filtrar con el df anterior los ataques desde el a침o 1990

df_paises_1900 = df_paises[df_paises['year'] >=1900.0]
df_paises_1900.head(2)
```

```python
plt.figure(figsize =(20,14))

fig = sns.countplot(x='country',hue = 'year', data = df_paises_1900)

fig.set(xlabel='Pais', ylabel = 'Ataques')

fig.set_title('Ataques por paises desde 1900')

fig.legend(bbox_to_anchor =(1,1),ncol = 4)

plt.show()
```

A partir de esta gr치fica podemos sacar las siguientes conclusiones:
- El aumento de la temperatura con el paso de los a침os y el cambio clim치tico, hace que las aguas sean m치s calidas y atraigan a m치s tiburones.
- M치s ataques de tiburones en el Atl치ntico que en el Pac칤fico debido a mayor n칰mero de turismo.
- La sequ칤a en Sud치frica podr칤a ser una variable que afecte a la disminuci칩n del os ataques en comparaci칩n con las aguas de Australia y USA
